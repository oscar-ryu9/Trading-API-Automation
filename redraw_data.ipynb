```python
def create_axes(screen, width, height):
    BLACK = (0, 0, 0)
    pygame.draw.line(screen, BLACK, (50, 50), (width - 50, 50), 2)  # Top line
    pygame.draw.line(screen, BLACK, (50, height - 50), (width - 50, height - 50), 2)  # Bottom line
    pygame.draw.line(screen, BLACK, (50, 50), (50, height - 50), 2)  # Left line
    pygame.draw.line(screen, BLACK, (width - 50, 50), (width - 50, height - 50), 2)  # Right line
```

```python
background_color = (22,26,37)


pygame.init()
screen_width = 2050
screen_height = 1050
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Pygame Window")

running = True
while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill(background_color)

    # Update the display
    pygame.display.flip()

    # Limit frame rate
    pygame.time.Clock().tick(60)

# Quit Pygame
pygame.quit()
```
